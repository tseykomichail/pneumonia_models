```python
import pandas as pd
train_labels=pd.read_csv('rsna-pneumonia-detection-challenge/stage_2_train_labels.csv')
train_labels.head()
```

```python
print(len(train_labels))
print(sum(train_labels['Target']==0))
print(sum(train_labels['Target']==1))
```

```python
import os
import shutil
import random


print(train_labels.loc[4]['patientId'])
print(train_labels.loc[5]['patientId'])
print(len(os.listdir('/content/rsna-pneumonia-detection-challenge/stage_2_train_images')))


train_dir = 'train'
val_dir = 'val'
test_dir = 'test'

normal_train_dir = os.path.join(train_dir, 'NORMAL')
pneumonia_train_dir = os.path.join(train_dir, 'PNEUMONIA')
normal_val_dir = os.path.join(val_dir, 'NORMAL')
pneumonia_val_dir = os.path.join(val_dir, 'PNEUMONIA')
normal_test_dir = os.path.join(test_dir, 'NORMAL')
pneumonia_test_dir = os.path.join(test_dir, 'PNEUMONIA')

source_dir='/content/rsna-pneumonia-detection-challenge/stage_2_train_images'

for i in range(len(train_labels)) :
  if train_labels.loc[i]['Target']==0 :
    src = os.path.join(source_dir, train_labels.loc[i]['patientId'] + '.dcm')
    if os.path.exists(src) :
      dst = os.path.join(normal_train_dir, train_labels.loc[i]['patientId']+ '.dcm')
      shutil.move(src, dst)
  else :
    src = os.path.join(source_dir, train_labels.loc[i]['patientId']+ '.dcm')
    if os.path.exists(src) :
      dst = os.path.join(pneumonia_train_dir, train_labels.loc[i]['patientId']+ '.dcm')
      shutil.move(src, dst)
```

```python

normal_images = os.listdir(normal_train_dir)
pneumonia_images = os.listdir(pneumonia_train_dir)
print('Тренировочная выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
print()
normal_val_dir = os.path.join(val_dir, 'NORMAL')
pneumonia_val_dir = os.path.join(val_dir, 'PNEUMONIA')

normal_images = os.listdir(normal_val_dir)
pneumonia_images = os.listdir(pneumonia_val_dir)

print('Валидационная выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
print()

normal_test_dir = os.path.join(test_dir, 'NORMAL')
pneumonia_test_dir = os.path.join(test_dir, 'PNEUMONIA')

normal_images = os.listdir(normal_test_dir)
pneumonia_images = os.listdir(pneumonia_test_dir)

print('Тестовая выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
```

```python
normal_train_dir = os.path.join(train_dir, 'NORMAL')
pneumonia_train_dir = os.path.join(train_dir, 'PNEUMONIA')
normal_images = os.listdir(normal_train_dir)
pneumonia_images = os.listdir(pneumonia_train_dir)
random.shuffle(normal_images)
random.shuffle(pneumonia_images)
print(normal_images)
```

```python
print(len(normal_images))
print(len(pneumonia_images))
coef=len(normal_images)/(len(normal_images)+len(pneumonia_images))
print(coef)
```

```python

num_images_to_move = 4000
num_normal=round(num_images_to_move*coef)
num_pneumonia=round(num_images_to_move*(1-coef))
for i  in range(num_normal):
    src = os.path.join(normal_train_dir, normal_images[i])
    dst = os.path.join(normal_val_dir, normal_images[i])
    shutil.move(src, dst)

for i  in range(num_pneumonia) :
    src = os.path.join(pneumonia_train_dir, pneumonia_images[i])
    dst = os.path.join(pneumonia_val_dir, pneumonia_images[i])
    shutil.move(src, dst)
```

```python

normal_images = os.listdir(normal_train_dir)
pneumonia_images = os.listdir(pneumonia_train_dir)
random.shuffle(normal_images)
random.shuffle(pneumonia_images)

num_images_to_move = 2000
num_normal=round(num_images_to_move*coef)
num_pneumonia=round(num_images_to_move*(1-coef))
for i  in range(num_normal):
    src = os.path.join(normal_train_dir, normal_images[i])
    dst = os.path.join(normal_test_dir, normal_images[i])
    shutil.move(src, dst)

for i  in range(num_pneumonia) :
    src = os.path.join(pneumonia_train_dir, pneumonia_images[i])
    dst = os.path.join(pneumonia_test_dir, pneumonia_images[i])
    shutil.move(src, dst)
```

```python

normal_images = os.listdir('train/NORMAL')
pneumonia_images = os.listdir('train/PNEUMONIA')
print('Тренировочная выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
print()

normal_images = os.listdir('val/NORMAL')
pneumonia_images = os.listdir('val/PNEUMONIA')

print('Валидационная выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
print()


normal_images = os.listdir('test/NORMAL')
pneumonia_images = os.listdir('test/PNEUMONIA')

print('Тестовая выборка:')
print('Без пневмонии ',  len(normal_images))
print('C пневмонией ',  len(pneumonia_images))
```
